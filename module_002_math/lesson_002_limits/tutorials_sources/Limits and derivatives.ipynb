# Пределы

Данна функция:
$f(x)=(1+x)^\frac{1}{x}$

Это функция не определена в точке 0. Но мы можем вычислить функцию в точки сколь угодно близкой к нулю

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def showPlot(x_start, x_max, step, callback, title='Function'):
    x = np.arange(x_start, x_max, step)
    y = callback(x)

    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()

def target_function(x):
    return (1+x)**(1/x)

x = 10.0
for i in range(1,10):
    print(x, ': ', target_function(x))
    x /= 10.0
    

showPlot(0.001, 10, 0.1, target_function)
```

$\lim\limits_{x \to 0}(1+x)^\frac{1}{x}=e$

Но не все функции имеют конечный предел в точке существуют и бесконечные пределы
$\lim\limits_{x \to 0}\frac{1}{x}=+\infty$

```python
def target_function(x):
    return 1/x

x = 10.0
for i in range(1,10):
    print(x, ': ', target_function(x))
    x /= 10.0
    
showPlot(0.1, 2, 0.01, target_function)
```

# Производные

Приращение функции:
![alt](./1.gif "Приращение функции")

$b = \frac{f(x+h)-f(x)}{h}$, где $b=tg\theta$

![alt](./2.gif "Тригонометрия")

Производная - это

$f'(x)=\lim\limits_{\Delta x \to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$

Гладкая функция - это такая функция, производная которой существует и непрерывна

![alt](./3.png "Таблица производных")

$f(x)=x^2; f'(x)=\lim\limits_{\Delta x \to 0} \frac{(x+\Delta x)^2 - x^2}{\Delta x}=\lim\limits_{\Delta x \to 0} \frac{x^2 + 2x\Delta x + \Delta x^2 - x^2}{\Delta x}=\lim\limits_{\Delta x \to 0} \frac{\Delta x(2x+\Delta x)}{\Delta x}=2x$

```python
x, y = symbols('x y')
diff(cos(x), x)
```

```python
e = x**3
dx = diff(e, x)
dx
```

```python
def value_of_expr(x_vector, symbol, expr):
    y = np.empty_like(x_vector)
    for index, element  in enumerate(x_vector):
        y[index] = (expr.subs(symbol, element))
    return y

showPlot(-10, 10, 0.01, lambda i: value_of_expr(i, x, e), title='Target')
showPlot(-10, 10, 0.01, lambda i: value_of_expr(i, x, dx), title='Diff')
```

# Производная сложной функции

Сложная функция - $g(x) = h(f(x))$, пример: $g(x)=sin(x+1), h(x)=sin(x), f(x)=1+x$

$df = f'(x)dx,\ где\ dx = \Delta x$

$f'(x)=\frac{df}{dx}$

Chain rule:

Дано:
$z=g(y), y=h(x)$

$\frac{dz}{dx}=\lim\limits_{\sigma x \to 0}\frac{\sigma z}{\sigma y} \frac{\sigma y}{\sigma x}=\lim\limits_{\sigma y \to 0}\frac{\sigma z}{\sigma y} \lim\limits_{\sigma x \to 0}\frac{\sigma y}{\sigma x}=\frac{dz}{dy} \frac{dy}{dx}$

Наш пример: $g(x)=sin(x+1), g'(x)=1*(cos(x+1))$

# Экстремумы

Экстремумы - это точки, для которых существует окрестность, где функция $f(x) >= f(x_0)\ или\ f(x) <= f(x_0),\ где\ x_0\ - \ экстремум,\ а\ x \in U$

![alt](./4.jpg "Экстремумы")

Необходимым условием экстремума, но не достаточным является то, что $f'(x_0)=0$

* $f'(x) > 0$ - функция строго возрастает
* $f'(x) < 0$ - функция строго убывает
* $f''(x) > 0$ - функция строго выпукла
* $f''(x) < 0$ - функция строго вогнута

Достаточным условием при выполнение необходимого является, то что вторая произовдная должна быть строго больше 0 или меньше

$f''(x) > 0 - минимум$

$f''(x) < 0 - максимум$

```python
e = .1*((log(x)/2.)**sin(x))

showPlot(3, 10, 0.1, lambda i: value_of_expr(i, x, e), title='Target')
```

```python
dx = diff(e, x)
dx
```

```python
showPlot(3, 10, 0.1, lambda i: value_of_expr(i, x, dx), title='Target')
```

```python
second_dex = dx = diff(e, x, x)
second_dex
```

```python
showPlot(3, 10, 0.1, lambda i: value_of_expr(i, x, second_dex), title='Target')
```
